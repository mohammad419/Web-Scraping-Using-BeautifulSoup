## Loading Required Libraries

```python
from bs4 import BeautifulSoup as bs
import requests
```

## Creating a variable for the HTML page to scrap data from and Requesting webpage to access its HTML page by right clicking the mouse and then clicking "Inspect the page" option 

```python
link = 'https://www.amazon.in/Acer-Nitro-Intel-Core-15-6-inch/product-reviews/B088FLYQJ9/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

### Creating an Object for traversing the html webpage 

```python
soup = bs(page.content,'html.parser')
```

### Scraping the names of the people who reviewed the product 
### For that you need the class and the tag associated with the data to do that you need to first rightclick the mouse on the data which you want to scrap and choose the option "inspect" to know the class and the tag of the data and then copy that
### Here span - "tag" and class_-'a-profile-name'

```python
names = soup.find_all('span',class_='a-profile-name')
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0)
cust_name.pop(0)
cust_name
```

### Scraping the title of the reviews of the product
### For that you need the class and the tag associated with the data to do that you need to first rightclick the mouse on the data which you want to scrap and choose the option "inspect" to know the class and the tag of the data and then copy that

```python
title = soup.find_all('a',class_='review-title-content')
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

### Scraping the rating given to the product
### For that you need the class and the tag associated with the data to do that you need to first rightclick the mouse on the data which you want to scrap and choose the option "inspect" to know the class and the tag of the data and then copy that

```python
rating = soup.find_all('i',class_='review-rating')
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate.pop(0)
rate.pop(0)
rate
```

### Scraping the reviews of the product
### For that you need the class and the tag associated with the data to do that you need to first rightclick the mouse on the data which you want to scrap and choose the option "inspect" to know the class and the tag of the data and then copy that

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content[:] = [reviews.replace('\n',"") for reviews in review_content]
review_content[:] = [reviews.replace('\xa0','') for reviews in review_content]
review_content
```

### Applying a tabular data structure

```python
import pandas as pd
df = pd.DataFrame()
df['Customer Name']=cust_name
df['Review title']=review_title
df['Ratings']=rate
df['Reviews']=review_content
df
```

### Exporting the dataset to a CSV file

```python
df.to_csv('Product_Reviews.csv', index=False)
```
